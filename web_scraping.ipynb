```python
# Import of libraries

from bs4 import BeautifulSoup
import requests
import pandas as pd
import json
import re
```

```python
# Using an API to not get 403 forbidden

response = requests.get(
  url='https://proxy.scrapeops.io/v1/',
  params={
      'api_key': '9d93d94a-bc84-455d-ac1c-59436d56fbd6',
      'url': 'https://www.terabyteshop.com.br/hardware/processadores/intel', 
  },
)

print('Response Body: ', response.content)
```

```python
# Configuration of beautifulsoup

soup = BeautifulSoup(response.content, 'html.parser')
```

```python
# Using beautifulsoup methods to get the required data from the website

prod_descriptions = soup.find_all('div', class_='commerce_columns_item_caption')
prod_prices = soup.find_all('div', class_='col-xs-12 col-sm-12 col-md-12')
```

```python
# Begin cleaning the data 

prod_description = []
for prod in prod_descriptions:
     prod_description.append(prod.get_text().replace('\n', ''))

final_prod_description = []
[final_prod_description.append(x) for x in prod_description if x not in final_prod_description]
     

prod_price = []
for price in prod_prices:
    prod_price.append(price.get_text().replace('\n', ''))


print(str(prod_price))
#print(str(final_prod_description))
```

```python
# Creating three different lists to append three types of descriptions

category = []
for i in prod_description:
    category.append(i[0:11])
#print(categoria)

brand = []
for i in prod_description:
    brand.append(i[12:17])
#print(marca)

model = []
for i in prod_description:
    model.append(i[18:36])
#print(modelo)
```

```python
# Slicing the lists to select only the necessary data

final_brand = brand[0:54]
final_model = model[0:54]
final_category = category[0:54]
```

```python
# Creating three different lists to append three payment options

full_price = []
for i in prod_price:
    full_price.append(i[1:15])

# IMPORTANT! In my country it's common to make installments, so I created these lists to append the payment options we get from the website
# Also you get an discount if you pay the full price on demand, that's why we have three payment options

price_on_demand = []
for i in prod_price:
    price_on_demand.append(i[20:34])

price_installments = []
for i in prod_price:
    price_installments.append(i[46:58])

#print(full_price)
#print(price_on_demand)
print(price_installments)
```

```python
# Cleaning the data

semifinal_full_price = []
for i in full_price:
    f = re.sub(r'[a-zA-Z:$:à]', '', i)
    semifinal_full_price.append("".join(f))

final_full_price = [i for i in semifinal_full_price if i.strip()]

final_full_price = final_full_price[0:54]

print(final_full_price)
```

```python
# Cleaning the data

semifinal_price_on_demand = []
for i in price_on_demand:
    f = re.sub(r'[a-zA-Z:$:à]', '', i)
    semifinal_price_on_demand.append("".join(f))

final_price_on_demand = [i for i in semifinal_price_on_demand if i.strip()]

final_price_on_demand = final_price_on_demand[0:54]

print(final_price_on_demand)
```

```python
# Cleaning the data

semifinal_price_installments = []
for i in price_installments:
    f = re.sub(r'[a-zA-Z:$:à]', '', i)
    semifinal_price_installments.append("".join(f))

final_price_installments = [i for i in semifinal_price_installments if i.strip()]

final_price_installments = final_price_installments[0:54]

print(final_price_installments)
```

```python
# Creating the DataFrame of the data we scraped, also creating id column

df = pd.DataFrame(list(zip(final_category, final_brand, final_model, final_full_price, final_price_on_demand, final_price_installments)), 
                           columns=['category', 'brand',  'model', 'full_price', 'price_on_demand', 'installments_12x'])

df["product_id"] = df.index + 1
```

```python
# Rearranging the columns for better organization and visualizing the DataFrame

cols = list(df.columns)
cols = [cols[-1]] + cols[:-1]
df = df[cols]
df
```

```python
df.to_csv("intel_prices_data.csv", index=False, encoding='utf-8')
```
